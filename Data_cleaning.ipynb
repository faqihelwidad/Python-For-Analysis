```python
import pandas as pd

df = pd.read_csv(r"dataset_penjualan_online_kotor.csv")

df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.head(10)
```

```python
df["Produk"].unique()
```

```python
df["Produk"] = df["Produk"].str.replace("Webacam","Webcam")
```

```python
df["Kota"].unique()
```

```python
p = {"Surabya" : "Surabaya",
      "Yogykarta" : "Yogyakarta",
     "Tangrang" : "Tanggerang",
     "Jakrta" : "Jakarta"}

df["Kota"] = df["Kota"].replace(p)
```

```python
df["Metode_Pembayaran"].unique()
```

```python
df["Kategori"].unique()
```

```python
df = df.drop(columns="Total")

df
```

```python
df["Total"] = df["Harga"] * df["Jumlah"]

df
```

```python
df["Tanggal"] = pd.to_datetime(df["Tanggal"])
```
