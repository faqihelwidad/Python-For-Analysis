```python
import pandas as pd

df = pd.read_csv(r"ecommerce_dirty_data_large.csv")

df.head(10)
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()

df.isnull().sum()
```

```python
df.head(5)
```

```python
df["Customer_Name"] = df["Customer_Name"].str.title()

df
```

```python
df["City"].unique()
```

```python
mapping_kota = {
    'jakarta': 'Jakarta', 'JAKARTA': 'Jakarta', 'Jakrta': 'Jakarta', 'JKT': 'Jakarta','Jkt': 'Jakarta',
    'bandung': 'Bandung', 'Bandug': 'Bandung',
    'surabaya': 'Surabaya', 'Sby': 'Surabaya',
    'bali': 'Bali',
    'medan': 'Medan', 'MEDAN': 'Medan', 'Medn': 'Medan'
}

df["City"] = (df["City"].str.strip().str.title().replace(mapping_kota))

df["City"].unique()
```

```python
df["Quantity"].unique()
```

```python
df["Quantity"] = df["Quantity"].replace("dua", 2)
df["Quantity"] = df["Quantity"].replace("Tiga", 3)

df["Quantity"].unique()
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
```

```python
df["Quantity"].unique()
```

```python
df.head()
```

```python
df["Price"].unique()
```

```python
df["Price"] = df["Price"].str.replace(".", "")
df["Price"] = df["Price"].replace("20rb", 20000)

df["Price"].unique()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df["Price"] = pd.to_numeric(df["Price"])
```

```python
df.dtypes
```

```python
df["Total"].unique()
```

```python
df["Total"] = df["Total"].str.replace(".", "")
df["Total"] = df["Total"].replace("error", None)
```

```python
df = df.dropna()
```

```python
df["Total"] = pd.to_numeric(df["Total"])
```

```python
df.head()
```

```python
import re

def valid_email(email):
  if pd.isna(email):
    return None
  email = email.strip().lower()
  pola = r'^[\w\.-]+@[\w\.-]+\.\w+$'
  return email if re.match(pola, email) else None

df["Email_Bersih"] = df["Email"].apply(valid_email)
```
